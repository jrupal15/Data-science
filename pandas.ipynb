```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

mm=pd.read_csv('titanic.csv')
mm.head(20)


#mm.describe()
```

```python
sur=mm[mm['Survived']==1]['Sex'].value_counts()
sur
```

```python
dead=mm[mm['Survived']==0]['Sex'].value_counts()
dead
```

```python
sur_dead=pd.DataFrame([sur,dead])
sur_dead.index=['survied','dead']
sur_dead
```

```python
sur_dead.plot(kind='bar',title='ratio of survive and death during titanic acident')
plt.show()
```

```python
plt.hist([mm[mm['Survived']==1]['Age'],mm[mm['Survived']==0]['Age']],stacked=True,color=['b','g'],bins=30,label=['Survived','dead'])
plt.xlabel('Age')
plt.ylabel('no. of passengers')
plt.legend()
plt.show()
```
